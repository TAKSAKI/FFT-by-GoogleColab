```
import cv2
import numpy as np
import matplotlib.pyplot as plt
src = cv2.imread(uploaded_file_name,0)
plt.imshow(src)
plt.gray()
```

```
mg_fft = np.fft.fft2(src)
fsrc = np.fft.fftshift(mg_fft)
plt.gray()
plt.imshow(np.abs(fsrc))
```

```
h,w=fsrc.shape
size=fsrc.shape
filter_matrix = np.zeros(size)
cy,cx=h/2,w/2
r=10
for y in range(0,h):
     for x in range(0,w):
            if pow((y-cy)**2,0.5)<r and pow((x-cx)**2,0.5)<r:
                filter_matrix[y][x] = 1
```

```

fimage = fsrc*filter_matrix
fimage = np.fft.fftshift(fimage) #shiftする
ifimage = np.fft.ifft2(fimage) #Inverse Fourier Transform
ifimage = ifimage.real
ifimage=np.uint8(ifimage)
```

```
from matplotlib import pyplot as plt

plt.imshow(ifimage)
```
